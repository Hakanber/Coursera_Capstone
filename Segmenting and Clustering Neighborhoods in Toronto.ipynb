1-As suggested, I use the following packages

```python
#Use pandas, or the BeautifulSoup package, or any other way you are comfortable with 
#to transform the data in the table on the Wikipedia page into the above pandas dataframe.
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
print("The packages are called in.")
```

2- Fetch the data from the link provided in the assignment

```python
my_url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
url_fetched = requests.get(my_url).text
fetched_data = BeautifulSoup(url_fetched, 'lxml')
table=fetched_data.find('table')
```

3- It is asked that the dataframe should consist of three columns: PostalCode, Borough, and Neighborhood, therefore:

```python
ColNames = ['Postalcode','Borough','Neighborhood']
my_df = pd.DataFrame(columns = ColNames)

for tr_cell in table.find_all('tr'):
    row_data=[]
    for td_cell in tr_cell.find_all('td'):
        row_data.append(td_cell.text.strip())
    if len(row_data)==3:
        my_df.loc[len(my_df)] = row_data

my_df.head()
```

4- Now, it's time to clean the dataset according to the assignment

```python
#Only process the cells that have an assigned borough. Ignore cells with a borough that is Not assigned.
my_df = my_df[my_df.Borough != 'Not assigned'] 

#Rows will be combined into one row with the neighborhoods separated with a comma.
my_df_grouped = my_df.groupby(['Postalcode','Borough'], sort=False).agg(', '.join)

#To seperate the columns, from 1 to 3.
my_df_grouped.reset_index(inplace=True)

#If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough.
my_df_grouped['Neighborhood'] = np.where(my_df_grouped['Neighborhood'] == 'Not assigned',my_df_grouped['Borough'], my_df_grouped['Neighborhood'])

my_df_grouped.rename(columns={'Postal Code':'PostalCode'},inplace=True) #Change the column name of "Postal Code" as shown "PostalCode"

my_df_grouped.head()
```

## TASK III

#### Explore and cluster the neighborhoods in Toronto

```python
!conda install -c conda-forge folium=0.5.0 --yes
import folium # plotting library
from bs4 import BeautifulSoup
from sklearn.cluster import KMeans
import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python

# set number of clusters
kclusters = 5

toronto_grouped_clustering = my_third_df.drop(['Postalcode','Borough','Neighborhood'],1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_
```

```python
# add clustering labels
my_third_df.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
my_third_df.head()
```

```python
# create map
map_clusters = folium.Map(location=[43.651070,-79.347015], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(my_third_df['Latitude'], my_third_df['Longitude'], my_third_df['Neighborhood'], my_third_df['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
